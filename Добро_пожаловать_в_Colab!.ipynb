```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# 10 ta mahsulot nomlari
mahsulotlar = ["Olma", "Banan", "Shaftoli", "Apelsin", "Uzum",
               "Anor", "Gilos", "Qovun", "Tarvuz", "Limon"]

# Narxlar va sotilgan miqdorlar
np.random.seed(42)
narxlar = np.random.randint(10000, 30000, 10)
sotilgan = np.random.randint(50, 200, 10)
df = pd.DataFrame({
    "Mahsulot": mahsulotlar,
    "Narx": narxlar,
    "Sotilgan_soni": sotilgan
})

# Umumiy daromad ustuni
df["Daromad"] = df["Narx"] * df["Sotilgan_soni"]

df
print("ðŸ“Š Umumiy statistik maâ€™lumotlar:")
print(df.describe())

print("\nðŸ’° Eng koâ€˜p daromad keltirgan mahsulot:")
print(df.loc[df["Daromad"].idxmax()])

print("\nðŸ“‰ Eng kam daromad keltirgan mahsulot:")
print(df.loc[df["Daromad"].idxmin()])
plt.figure(figsize=(9,5))
plt.bar(df["Mahsulot"], df["Daromad"], color="skyblue", edgecolor="black")
plt.title("Mahsulotlar boâ€˜yicha umumiy daromad", fontsize=14)
plt.xlabel("Mahsulotlar")
plt.ylabel("Daromad (soâ€˜m)")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
