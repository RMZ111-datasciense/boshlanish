```
# === 1. KERAKLI KUTUBXONALAR ===
import pandas as pd
import matplotlib.pyplot as plt

# === 2. MA'LUMOTLAR YARATISH ===
data = {
    "Viloyat": ["Toshkent", "Samarqand", "Fargâ€˜ona", "Namangan", "Andijon",
                 "Buxoro", "Navoiy", "Xorazm", "Qashqadaryo", "Surxondaryo"],
    "Aholi (mln)": [3.0, 2.2, 2.1, 1.9, 2.0, 1.8, 1.1, 1.7, 2.3, 2.5],
    "Oâ€˜rtacha daromad (ming soâ€˜m)": [4200, 3500, 3300, 3100, 3000, 3700, 4800, 3400, 3200, 2900],
    "Ishsizlik (%)": [6.0, 7.5, 8.1, 8.3, 8.5, 7.0, 5.2, 8.0, 8.4, 8.7]
}

df = pd.DataFrame(data)

# === 3. ASOSIY TAHLIL ===
print("ðŸ’° Eng yuqori daromadli viloyat:")
print(df.loc[df["Oâ€˜rtacha daromad (ming soâ€˜m)"].idxmax()])

print("\nðŸ˜• Eng past daromadli viloyat:")
print(df.loc[df["Oâ€˜rtacha daromad (ming soâ€˜m)"].idxmin()])

# === 4. KORRELYATSIYA TAHLILI ===
corr = df["Aholi (mln)"].corr(df["Oâ€˜rtacha daromad (ming soâ€˜m)"])
print(f"\nðŸ“ˆ Aholi soni va daromad oâ€˜rtasidagi bogâ€˜liqlik darajasi (korrelyatsiya): {corr:.2f}")

# Jadvalni chiqaramiz
df
# === DAROMADLAR GRAFIGI ===
plt.figure(figsize=(10,6))
plt.bar(df["Viloyat"], df["Oâ€˜rtacha daromad (ming soâ€˜m)"], color='skyblue')
plt.title("Viloyatlar boâ€˜yicha oâ€˜rtacha daromad")
plt.xlabel("Viloyat")
plt.ylabel("Daromad (ming soâ€˜m)")
plt.xticks(rotation=45)
plt.show()

# === AHOLI SONI VA DAROMADNING SCATTER GRAFIGI ===
plt.figure(figsize=(8,6))
plt.scatter(df["Aholi (mln)"], df["Oâ€˜rtacha daromad (ming soâ€˜m)"], color='orange')
plt.title("Aholi soni va daromad oâ€˜rtasidagi bogâ€˜liqlik")
plt.xlabel("Aholi (mln)")
plt.ylabel("Daromad (ming soâ€˜m)")
plt.grid(True)
plt.show()
print("ðŸ§¾ Xulosa:")
print("""
- Eng boy viloyat: Navoiy
- Eng past daromadli viloyat: Surxondaryo
- Korrelyatsiya musbat (aholi koâ€˜p boâ€˜lsa, daromad koâ€˜tarilmoqda)
- Grafiklar orqali koâ€˜ramizki, Toshkent va Navoiy iqtisodiy jihatdan yuqorida.
""")
```
